## 3. Run your first AI signal scan

The bot downloads free data, creates indicators, trains a small AI model, then prints signals.


````python
import os
os.environ['SYMBOLS'] = 'EURUSD,GBPUSD,USDJPY,AUDUSD,USDCAD,USDCHF,NZDUSD'
os.environ['TIMEFRAME'] = '1h'
os.environ['PERIOD'] = '180d'
os.environ['MIN_SIGNAL_STRENGTH'] = '0.62'
os.environ['SIGNAL_OUTPUT'] = 'latest_signals.csv'

from forex_ai_bot import scan_symbols
signals = scan_symbols(['EURUSD','GBPUSD','USDJPY'], timeframe='1h', period='180d', output_path='latest_signals.csv')

import pandas as pd
pd.DataFrame(signals)
````
